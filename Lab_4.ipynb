```python
from GoogleNews import GoogleNews
import pandas as pd
```

```python
googlenews = GoogleNews(lang='en')
```

```python
def get_news(target, num_pages=1):
    googlenews.clear()
    googlenews.search(target)
    articles = []
    for i in range(1, num_pages + 1):
        googlenews.getpage(i)
        articles.extend(googlenews.result())
    return articles
```

```python
def create_dataset(news_df, filename):
    df = pd.DataFrame(news_df)
    df.to_csv(filename, index=False)
```

```python
topics = ['curling', 'tanzania', 'panda', 'airbus']
```

```python
all_articles = []
for topic in topics:
    articles = get_news(topic, num_pages=2)
    for article in articles:
        article['topic'] = topic
    all_articles.extend(articles)
    topic_articles = [article for article in all_articles if article['topic'] == topic]
    create_dataset(topic_articles, f'{topic}_news.csv')
```

```python
for topic in topics:
    topic_articles = [article for article in all_articles if article['topic'] == topic]
    print(f"Topic: {topic}")
    topic_articles = pd.DataFrame(topic_articles)
    topic_articles = topic_articles [:5]
    print(display(pd.DataFrame(topic_articles)))
```

```python
publishers = ['BBC', 'NBC News', 'Al Jazeera', 'Euronews']
```

```python
all_articles = []
for publisher in publishers:
    articles = get_news(publisher, num_pages=2)
    for article in articles:
        article['publisher'] = publisher
    all_articles.extend(articles)
    publisher_articles = [article for article in all_articles if article['publisher'] == publisher]
    create_dataset(publisher_articles, f'{publisher}_news.csv')
```

```python
for publisher in publishers:
    publisher_articles = [article for article in all_articles if article['publisher'] == publisher]
    print(f"Publisher: {publisher}")
    publisher_articles = pd.DataFrame(publisher_articles)
    publisher_articles = publisher_articles [:5]
    print(display(pd.DataFrame(publisher_articles)))
```

```python
dates = ['05/25/2024', '05/25/2023', '05/25/2022']
```

```python
all_articles = []

for date in dates:
    articles = get_news(date, num_pages=2)
    for article in articles:
        article['date'] = date
    all_articles.extend(articles)
    date_articles = [article for article in all_articles if article['date'] == date]
    create_dataset(date_articles, f'news_{date.replace("/", "-")}.csv')
```

```python
for date in dates:
    date_articles = [article for article in all_articles if article['date'] == date]
    print(f"Date: {date}")
    date_articles = pd.DataFrame(date_articles)
    date_articles = date_articles [:5]
    print(display(pd.DataFrame(date_articles)))
```
